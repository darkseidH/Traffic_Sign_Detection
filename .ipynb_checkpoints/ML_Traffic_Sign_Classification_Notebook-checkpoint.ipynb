## Chargement des données de l'entrainement

```python
# Définition du chemin de base où se trouvent les données pour le projet de classification de panneaux de signalisation.
base_dir = "./traffic_Data/DATA/"

# Liste des noms des dossiers représentant différentes classes de panneaux. Chaque dossier contient des images associées à une classe spécifique.
class_folder_names = ['0', '4', '6', '11', '12', '14', '17']
```

```python
# Initialisation d'une liste vide pour stocker les chemins des images.
image_paths = []

# Boucle à travers la liste des noms de dossiers de classe spécifiée.
for cls in class_folder_names:
    # Boucle à travers chaque fichier dans le dossier de la classe.
    for file_name in os.listdir(os.path.join(base_dir, cls)):
        # Vérifie si le fichier est une image PNG.
        if file_name.split('.')[-1] == 'png':
            # Ajoute le chemin complet de l'image à la liste image_paths.
            image_paths.append(os.path.join(base_dir, cls, file_name))

# Affichage du nombre total d'images collectées.
print("A total images = ", len(image_paths))
print("-------------------------------")
```

```python
# Initialisation d'une liste vide pour stocker les noms des classes.
classes = []

# Boucle sur chaque chemin d'image dans la liste image_paths.
for image_path in image_paths:
    # Extraction du nom de la classe à partir du chemin de l'image.
    # Le chemin est divisé en segments et le segment avant le dernier ([-2]) correspond au nom du dossier de la classe.
    classes.append(image_path.split('/')[-2])
```

## Chargement des données de test

```python
# Initialisation d'une liste vide pour stocker les chemins des images de test.
image_test_paths = []

# Définition du chemin du répertoire contenant les images de test.
test_dir = "./traffic_Data/NEW_TEST/"

# Initialisation d'une liste vide pour stocker les classes des images de test.
y_test_classes = []

# Boucle à travers la liste des noms des dossiers de classe spécifiée pour le test.
for cls in class_folder_names:
    # Boucle à travers chaque fichier dans le dossier de classe spécifié pour le test.
    for file_name in os.listdir(os.path.join(test_dir, cls)):
        # Ajoute la classe à la liste y_test_classes.
        y_test_classes.append(cls)
        # Vérifie si le fichier est une image PNG.
        if file_name.split('.')[-1] == 'png':
            # Ajoute le chemin complet de l'image de test à la liste image_test_paths.
            image_test_paths.append(os.path.join(test_dir, cls, file_name))

# Affichage du nombre total d'images de test collectées.
print("A total images = ", len(image_test_paths))
print("-------------------------------")
```

# Prétraitement des données : Said ElHafyani et Siham Demami

## Prétraitement des images avant de les passés à l'éxtraction des caractéristiques

### Données de l'entrainement

```python
from rembg import remove

SIZE = 256  # Définition de la taille à laquelle toutes les images seront redimensionnées

inputs = []  # Initialisation d'une liste pour stocker les images traitées

# Boucle sur chaque chemin d'image dans la liste des chemins d'images
for img_path in tqdm(image_paths):
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)  # Lecture de l'image en couleurs
    img = cv2.resize(img, (SIZE, SIZE))  # Redimensionnement de l'image à la taille spécifiée
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)  # Conversion de l'espace de couleur de RGB à BGR pour le traitement
    result = remove(img)  # Application de la fonction de suppression d'arrière-plan
    result = cv2.resize(result, (SIZE, SIZE))  # Redimensionnement du résultat pour garantir l'uniformité
    result = cv2.cvtColor(result, cv2.COLOR_RGB2BGR)  # Conversion de l'espace de couleur de retour en BGR
    result = result / 255.0  # Normalisation des valeurs des pixels pour qu'elles soient entre 0 et 1
    inputs.append(result)  # Ajout de l'image traitée à la liste des entrées

X_train = np.array(inputs)  # Conversion de la liste en tableau NumPy pour l'utilisation dans les modèles de machine learning
```

### Données de test

```python
# Réinitialisation de la liste des entrées pour stocker les images de test traitées.
inputs = []

# Boucle sur chaque chemin d'image dans la liste des chemins d'images de test
for img_path in tqdm(image_test_paths):
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)  # Lecture de l'image en couleurs
    img = cv2.resize(img, (SIZE, SIZE))  # Redimensionnement de l'image à la taille spécifiée
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)  # Conversion de l'espace de couleur de RGB à BGR
    result = remove(img)  # Application de la fonction de suppression d'arrière-plan
    result = cv2.resize(result, (SIZE, SIZE))  # Redimensionnement du résultat pour garantir l'uniformité
    result = cv2.cvtColor(result, cv2.COLOR_RGB2BGR)  # Conversion de l'espace de couleur de retour en BGR
    result = result / 255.0  # Normalisation des valeurs des pixels pour qu'elles soient entre 0 et 1
    inputs.append(result)  # Ajout de l'image traitée à la liste des entrées

X_test = np.array(inputs)  # Conversion de la liste en tableau NumPy pour l'utilisation dans les tests de modèles de machine learning
```

### Transformation des class de l'entrainement et du test

```python
le = LabelEncoder()  # Création d'une instance de LabelEncoder

# Transformation des classes de formation en valeurs numériques et conversion en tableau numpy.
# La méthode fit_transform trouve tous les labels uniques et les assigne à des valeurs numériques.
y_train = le.fit_transform(classes)
y_train = np.array(y_train)  # Conversion de la liste en tableau numpy pour une manipulation plus efficace

# Transformation des classes de test en utilisant le même encodeur pour garantir la cohérence des labels entre les ensembles de données.
# Ici, il est plus typique d'utiliser la méthode transform au lieu de fit_transform pour les données de test pour utiliser le même encodage que pour y_train.
y_test = le.transform(y_test_classes)
y_test = np.array(y_test)  # Conversion en tableau numpy
```

### Chargement de model d'éxtarction des caractéristique

```python
# Création d'une instance du modèle VGG16 avec les poids pré-entraînés sur ImageNet.
# include_top=False signifie que la partie supérieure (les couches entièrement connectées) du réseau n'est pas incluse,
# ce qui permet de personnaliser le modèle pour des tailles d'entrée spécifiques et potentiellement pour de nouvelles classes.
# input_shape=(SIZE, SIZE, 3) définit la forme des images d'entrée que le modèle attend, ici des images de 256x256 avec 3 canaux de couleur (RGB).
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(SIZE, SIZE, 3))
```

```python
# Affichage d'un résumé de l'architecture du modèle VGG16.
# Ce résumé inclut les détails de toutes les couches dans le modèle, les formes des tenseurs de sortie après chaque couche,
# et le nombre total de paramètres dans le modèle. Cela est utile pour comprendre la structure du modèle et pour déboguer l'architecture du réseau.
base_model.summary()
```

```python
# Parcours de toutes les couches du modèle de base VGG16.
for layer in base_model.layers:
    layer.trainable = False  # Désactivation de l'entraînement pour chaque couche.
    # Cela verrouille les poids des couches pré-entraînées pour qu'ils ne soient pas modifiés lors de l'entraînement,
    # car nous voulons garder les caractéristiques apprises par le modèle sur ImageNet.

# Affichage du résumé du modèle après avoir marqué les couches comme non entraînables.
# Cela nous permet de vérifier que les couches sont effectivement marquées comme non entraînables (voir la colonne 'Trainable' dans le résumé).
base_model.summary()
```

### Extraction des caractéristique depuis les données de l'entrainement

```python
# Utilisation du modèle VGG16 comme un extracteur de caractéristiques.
# La méthode 'predict' traite les images d'entrée (X_train) et extrait les caractéristiques à la dernière couche de convolution.
feature_extractor = base_model.predict(X_train)

# Redimensionnement des caractéristiques extraites pour qu'elles soient compatibles avec des modèles de classification tels que Random Forest.
# Le redimensionnement est nécessaire car les modèles de machine learning classiques attendent des entrées en 2D (échantillons, caractéristiques),
# alors que les sorties des couches convolutives sont en 4D (échantillons, hauteur, largeur, canaux).
features = feature_extractor.reshape(feature_extractor.shape[0], -1)

# Stockage des caractéristiques redimensionnées dans X_for_RF, qui sera utilisé comme entrée pour le modèle de Random Forest.
X_for_RF = features
```

### Extraction des caractéristique depuis les données de test

```python
# Utilisation du modèle VGG16 comme un extracteur de caractéristiques pour l'ensemble de données de test.
# La méthode 'predict' traite les images d'entrée (X_test) et extrait les caractéristiques importantes à partir des couches convolutives.
feature_extractor = base_model.predict(X_test)

# Redimensionnement des caractéristiques extraites pour les adapter au format attendu par les modèles de machine learning traditionnels.
# Le reshape en 2D est nécessaire car les sorties du modèle VGG16 sont en 4D (échantillons, hauteur, largeur, canaux).
features = feature_extractor.reshape(feature_extractor.shape[0], -1)

# Stockage des caractéristiques redimensionnées dans X_for_RF_test, qui sera utilisé comme entrée pour tester le modèle de Random Forest.
X_for_RF_test = features
```

# Entrainement avec Un model de ML : Hamza Chaqchaq

```python
# Installation de la bibliothèque 'lazypredict' via pip. Lazypredict aide à construire rapidement de nombreux modèles de base
# pour évaluer leurs performances sans configuration manuelle approfondie. C'est un outil utile pour la sélection préliminaire de modèles
# en fournissant un aperçu des performances de divers algorithmes sur un jeu de données spécifique.
! pip install lazypredict
```

```python
# Importation de LazyClassifier depuis lazypredict.Supervised pour l'automatisation du processus de modélisation.
from lazypredict.Supervised import LazyClassifier

# Création d'une instance de LazyClassifier.
# `verbose=0` désactive les messages de progression, `ignore_warnings=True` permet d'ignorer les avertissements lors de l'entraînement des modèles,
# et `custom_metric=None` signifie qu'aucune métrique personnalisée n'est utilisée pour l'évaluation des modèles.
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)

# Ajustement et évaluation de plusieurs modèles de classification.
# `X_for_RF` sont les caractéristiques d'entraînement et `X_for_RF_test` les caractéristiques de test,
# `y_train` et `y_test` sont les étiquettes d'entraînement et de test respectivement.
# Cette méthode retourne deux DataFrames: `models` contenant les métriques de performance de chaque modèle,
# et `predictions` qui pourrait contenir les prédictions de chaque modèle (selon la version de lazypredict utilisée).
models, predictions = clf.fit(X_for_RF, X_for_RF_test, y_train, y_test)
# Nb : Il prend beaucoup de temps pour s'éxécuter
```

## Entrainement et evaluation du model avec les données d'entrainement et test  : Hamza Chaqchaq

```python
# Importation des bibliothèques et modules nécessaires
from sklearn.linear_model import RidgeClassifierCV
#Le modèle RidgeClassifierCV utilise la régularisation Ridge pour la classification.
#Le paramètre alphas spécifie les niveaux de régularisation à tester.
#Ici, avec les valeurs 0.1, 1.0, et 10.0, le modèle évaluera l'impact de trois différents degrés de pénalité sur
#la complexité du modèle et choisira le meilleur pour prévenir le surajustement et optimiser la performance.
model = RidgeClassifierCV(alphas=(0.1, 1.0, 10.0))
```

```python

# Entraînement du modèle RidgeClassifierCV sur l'ensemble d'entraînement.
# X_for_RF représente les caractéristiques d'entraînement et y_train les étiquettes associées.
model.fit(X_for_RF, y_train)

# Utilisation du modèle entraîné pour prédire les étiquettes de l'ensemble de test.
predictions = model.predict(X_for_RF_test)

# Calcul de l'exactitude de la prédiction en comparant les étiquettes prédites avec les véritables étiquettes de test.
accuracy = accuracy_score(y_test, predictions)

# Affichage de l'exactitude du modèle pour évaluer sa performance.
print(f"RidgeClassifierCV, accuracy: {accuracy}")
```

```python
# Récupération des coefficients du modèle après l'entraînement.
# Les coefficients reflètent l'importance et le signe de chaque caractéristique pour la prédiction du modèle.
coefficients = model.coef_.ravel()

# Affichage de la dimension du vecteur de coefficients.
# Cela nous indique le nombre de caractéristiques que le modèle a utilisées pour faire ses prédictions.
print(coefficients.shape)
```

```python
# Vérification si le tableau de coefficients n'est pas vide.
if coefficients is not None:
    # Création d'un histogramme des valeurs des coefficients pour analyser leur distribution.
    # 'nbins' définit le nombre de barres dans l'histogramme, permettant une granularité fine de la distribution.
    fig = px.histogram(x=coefficients, nbins=50, labels={'x': 'Coefficient Value'}, title='Distribution of Coefficients (Weights)')

    # Mise à jour de la mise en page pour réduire l'espacement entre les barres, rendant l'histogramme plus dense.
    fig.update_layout(bargap=0.1)

    # Mise à jour de la transparence des traces pour améliorer la lisibilité de l'histogramme.
    fig.update_traces(opacity=0.7)

    # Affichage de l'histogramme.
    fig.show()
```

# Rapport sur les résultats : Nouhayla Chouti

```python
# Génération d'un rapport de classification en utilisant les étiquettes réelles et les prédictions.
# Le rapport inclut la précision, le rappel et le score F1 pour chaque classe, ainsi que les moyennes pondérées.
report = classification_report(y_test, predictions)

# Affichage du rapport de classification pour analyser la performance du modèle.
print(report)
```

```python
# Calcul de la matrice de confusion entre les étiquettes réelles et les étiquettes prédites.
cm = confusion_matrix(y_test, predictions)

# Création d'une nouvelle figure avec matplotlib, avec une taille spécifiée pour une meilleure lisibilité.
plt.figure(figsize=(6, 4))

# Utilisation de seaborn pour créer un heatmap de la matrice de confusion.
# 'annot=True' pour annoter chaque cellule avec le nombre de prédictions,
# 'fmt='d'' pour formater les nombres en entiers,
# 'cmap='Blues'' pour utiliser une palette de couleurs bleues qui aide à visualiser les gradients de densité.
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=True, yticklabels=True)

# Ajout d'un titre avec le nom du meilleur modèle, basé sur l'exactitude.
plt.title('Confusion Matrix for the Best Model based on Accuracy (RidgeClassifierCV)')

# Étiquetage des axes pour clarifier quels sont les prévus versus les réels.
plt.xlabel('Predicted')
plt.ylabel('Actual')

# Affichage du graphique.
plt.show()
```
